```python
import numpy as np
import pandas as pd
import seaborn as sns

# Load penguins dataset
penguins = sns.load_dataset("penguins")

# Separate the 'species' column for targeting
y = penguins['species']

# Turn it into numeric data
mapping = {'Adelie': 2, 'Chinstrap': 1, 'Gentoo': 0}
y_numeric = np.array([mapping[val] for val in y])

# Add it to the dataset
penguins['target'] = y_numeric

# Plot pairwise relationships with sns.pairplot
palette = sns.color_palette(n_colors=3)
sns.pairplot(penguins, hue='target', palette={0: palette[0], 1: palette[1], 2: palette[2]})
```

```python
# Convert the dataset from multiclass to binary
y_binary = np.where(y_numeric == 0, 0, 1)
penguins['target'] = y_binary

# Plot again
sns.pairplot(penguins, hue='target')
```
